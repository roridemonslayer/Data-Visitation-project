```python
math_skewness = df['math score'].skew()
reading_skewness = df['reading score'].skew()
writing_skewness = df['writing score'].skew()
print('Skewness of Math Score:',math_skewness)
print('Skewness of Reading Score:',reading_skewness)
print('Skewness of Writing Score:',writing_skewness)
```

```python
colors = ['#f673a9', '#7ec8d4', '#ece5b4', '#9a4496']
plt.figure(figsize=(18, 12))

# Reading Score by Gender
plt.subplot(2, 3, 1)
sns.histplot(df[df['gender'] == 'female']['reading score'], kde=True, palette='husl', label='Female', bins=10)
sns.histplot(df[df['gender'] == 'male']['reading score'], kde=True,palette='husl' , label='Male', bins=10)
plt.title(f'Reading Score by Gender\nSkewness: {reading_skewness:.2f}')
plt.xlabel('Reading Score')
plt.legend()

# Reading Score by Race/Ethnicity
plt.subplot(2, 3, 2)
for group in df['race/ethnicity'].unique():
    sns.histplot(df[df['race/ethnicity'] == group]['reading score'], kde=True,palette='husl', label=group, bins=10)
plt.title('Reading Score by Race/Ethnicity')
plt.xlabel('Reading Score')
plt.legend()

# Reading Score by Parental Level of Education
plt.subplot(2, 3, 3)
for level in df['parental level of education'].unique():
    sns.histplot(df[df['parental level of education'] == level]['reading score'], kde=True,palette='husl', label=level, bins=10)
plt.title('Reading Score by Parental Level of Education')
plt.xlabel('Reading Score')
plt.legend()
plt.xticks(rotation=45)

# Reading Score by Lunch
plt.subplot(2, 3, 4)
sns.histplot(df[df['lunch'] == 'standard']['reading score'], kde=True,palette='husl', label='Standard', bins=10)
sns.histplot(df[df['lunch'] == 'free/reduced']['reading score'], kde=True, palette='husl', label='Free/Reduced', bins=10)
plt.title('Reading Score by Lunch')
plt.xlabel('Reading Score')
plt.legend()

# Reading Score by Test Preparation Course
plt.subplot(2, 3, 5)
sns.histplot(df[df['test preparation course'] == 'none']['reading score'], kde=True,palette='husl', label='None', bins=10)
sns.histplot(df[df['test preparation course'] == 'completed']['reading score'], kde=True,palette='husl', label='Completed', bins=10)
plt.title('Reading Score by Test Preparation Course')
plt.xlabel('Reading Score')
plt.legend()

plt.tight_layout()
plt.show()
```

```python

colors = ['#f673a9', '#7ec8d4', '#ece5b4', '#9a4496']
bins = [0, 60, 70, 100]  # Low (0-59), Medium (60-69), High (70-100)


df['math_score_category'] = pd.cut(df['math score'], bins=bins, labels=['Low', 'Medium', 'High'])
df['reading_score_category'] = pd.cut(df['reading score'], bins=bins, labels=['Low', 'Medium', 'High'])
df['writing_score_category'] = pd.cut(df['writing score'], bins=bins, labels=['Low', 'Medium', 'High'])

plt.figure(figsize=(18, 6))

# Pie chart for the math stuff
plt.subplot(1, 3, 1)
math_counts = df['math_score_category'].value_counts()
plt.pie(math_counts, labels=math_counts.index, autopct='%1.1f%%', colors=colors)
plt.title('Math Score Distribution')

# Pie chart for the reading stuff
plt.subplot(1, 3, 2)
reading_counts = df['reading_score_category'].value_counts()
plt.pie(reading_counts, labels=reading_counts.index, autopct='%1.1f%%', colors=colors)
plt.title('Reading Score Distribution')

# Pie chart for writing stuff
plt.subplot(1, 3, 3)
writing_counts = df['writing_score_category'].value_counts()
plt.pie(writing_counts, labels=writing_counts.index, autopct='%1.1f%%', colors=colors)
plt.title('Writing Score Distribution')

plt.tight_layout()
plt.show()
# what i mena by high low and medium is to show what percentage scored high low and mediuam on all sections.
```
